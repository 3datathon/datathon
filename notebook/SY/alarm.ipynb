### Ìå®ÌÑ¥ ÌååÏïÖ

```python
center_indices = [275477, 324761]

window = 10  # Ï†ÑÌõÑ 5ÌñâÏî© (Ï¥ù 11Ìñâ)

# Ï£ºÏöî ÏÑºÏÑú
selected_sensors = [
    'STATUS', 'KO1_MOTOR_SET_FREQ', 'CUTTING_SET_FREQ', 'MAIN_MOTOR_CURR', 'MAIN_MOTOR_SET_FREQ', 'MAIN_MOTOR_RPM'
]

pattern_list = []

for idx in center_indices:
    start_idx = max(0, idx - window)
    end_idx = min(len(df), idx + window + 1)
    
    slice_df = df.iloc[start_idx:end_idx].copy()
    slice_df['ALARM_CENTER_IDX'] = idx
    slice_df['RELATIVE_IDX'] = slice_df.index - idx
    pattern_list.append(slice_df)

alarm_patterns = pd.concat(pattern_list, ignore_index=True)
```

```python
for sensor in selected_sensors:
    if sensor not in alarm_patterns.columns:
        continue
    
    for idx in center_indices:
        temp = alarm_patterns[alarm_patterns['ALARM_CENTER_IDX'] == idx]
        plt.plot(temp['RELATIVE_IDX'], temp[sensor], label=f'Alarm@{idx}', alpha=0.7)
    
    plt.axvline(0, color='red', linestyle='--', label='Alarm Point')
    plt.title(f'{sensor} - ÏïåÎûå Ï†ÑÌõÑ Ìå®ÌÑ¥')
    plt.xlabel('Ìñâ ÏÉÅÎåÄ ÏúÑÏπò')
    plt.ylabel(sensor)
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()
```
